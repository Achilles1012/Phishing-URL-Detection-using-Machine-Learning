```python
# Assuming 'url' is your DataFrame and 'columns_to_remove' is a list of column names you want to remove
columns_to_remove = ['tld_params','extension','comma_url']  # Example list of columns to remove

# Removing the specified columns
url.drop(columns=columns_to_remove, inplace=True)
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Identify boolean columns
bool_columns = url.select_dtypes(include=bool).columns

# Exclude boolean columns and the column named 'URL'
columns_to_change = [col for col in url.columns if col not in bool_columns and col != 'URL']

# Change data types of selected columns
url[columns_to_change] = url[columns_to_change].astype('float32')  # Change 'new_data_type' to the desired data type

```

```python
url.dtypes
```

```python
X = url.iloc[:, :-2]
Y = url.iloc[:, -2]
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest=tts(X,Y,test_size=0.25,random_state=6)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, accuracy_score, confusion_matrix
```

```python
scores=[]
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Define a range of values for k
k_values = range(1, 21)

# Initialize an empty list to store accuracy scores
accuracy_scores = []

# Iterate over each value of k
for k in k_values:
    # Instantiate the KNN classifier with the current value of k
    knn_model = KNeighborsClassifier(n_neighbors=k)
    
    # Train the model
    knn_model.fit(xtrain, ytrain)
    
    # Make predictions on the test set
    knn_predictions = knn_model.predict(xtest)
    
    # Calculate accuracy and append to the list
    accuracy = accuracy_score(ytest, knn_predictions)
    accuracy_scores.append(accuracy)
    
    # Print the accuracy for the current value of k
    print(f"k={k}: Accuracy={accuracy}")

# Print the list of accuracy scores
print("\nAccuracy Scores:", accuracy_scores)
```

```python
from sklearn.naive_bayes import GaussianNB

# Instantiate the Gaussian Naive Bayes classifier
nb_model = GaussianNB()

# Train the model
nb_model.fit(xtrain, ytrain)

# Make predictions on the test set
nb_predictions = nb_model.predict(xtest)

# Evaluate the model
nb_accuracy = accuracy_score(ytest, nb_predictions)
print("Accuracy:", nb_accuracy)
```

```python
precision_nb = precision_score(ytest,nb_predictions )
recall_nb = recall_score(ytest, nb_predictions)
f1_nb = f1_score(ytest, nb_predictions)

print("Precision:", precision_nb)
print("Recall:", recall_nb)
print("F1-score:", f1_nb)
print(classification_report(ytest,nb_predictions))
scores.append([nb_accuracy,precision_nb,recall_nb,f1_nb])
```

```python
from sklearn.linear_model import LogisticRegression

# Assuming you have already split your data into training and testing sets: xtrain, xtest, ytrain, ytest

# Create a logistic regression model
logistic_regression = LogisticRegression()

# Train the model on the training data
logistic_regression.fit(xtrain, ytrain)

# Make predictions on the test data
log_predictions = logistic_regression.predict(xtest)

# Evaluate the model's accuracy
log_accuracy = accuracy_score(ytest, log_predictions)
print("Accuracy:", log_accuracy)
```
